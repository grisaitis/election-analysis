```python
import tabula
```

```python
dfs = tabula.read_pdf("./orange_precinct_2020.pdf", pages=list(range(1, 56)))
```

```python
import pandas as pd
```

```python
df = pd.concat(dfs)
df = df.iloc[:-5]
df
```

```python
df.loc[:, "countypct"] = df["Precinct - Contest"].apply(lambda text: "ORA" + text[-3:])
df
```

```python
orange_2020 = df.pivot(
    index="countypct",
    columns="Candidate",
    values="Total Votes"
)

orange_2020
```

```python
orange_2020.to_csv("./orange_2020_from_pdf.csv")
```
